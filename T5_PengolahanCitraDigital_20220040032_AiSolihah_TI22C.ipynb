```python
import imageio.v3 as image
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar dari path
path = "/content/Pemandangan.jpeg"
my_image = image.imread(path)

# Mengubah gambar menjadi grayscale
gray_image = np.mean(my_image, axis=2).astype(np.uint8)

# Fungsi untuk menyesuaikan kecerahan
def brightness(img, factor):
    img_result = img.astype(np.float32)
    img_result = img_result + factor
    img_result = np.clip(img_result, 0, 255)
    return img_result.astype(np.uint8)

# Fungsi untuk menyesuaikan kontras
def contrast(img, factor):
    mean = 128
    img_result = img.astype(np.float32)
    img_result = mean + factor * (img_result - mean)
    img_result = np.clip(img_result, 0, 255)
    return img_result.astype(np.uint8)

# Aplikasi brightness dan contrast pada gambar berwarna dan grayscale
bright_color_image = brightness(my_image, 50)
contrast_color_image = contrast(my_image, 2)
bright_gray_image = brightness(gray_image, 50)
contrast_gray_image = contrast(gray_image, 2)

# Menampilkan gambar asli, brightness, dan contrast pada citra berwarna dan grayscale
plt.figure(figsize=(12, 8))

# Citra Berwarna
plt.subplot(2, 3, 1)
plt.imshow(my_image, origin="upper")
plt.title("Gambar Asli (Berwarna)")

plt.subplot(2, 3, 2)
plt.imshow(bright_color_image, origin="upper")
plt.title("Brightness (Berwarna)")

plt.subplot(2, 3, 3)
plt.imshow(contrast_color_image, origin="upper")
plt.title("Contrast (Berwarna)")

# Citra Grayscale
plt.subplot(2, 3, 4)
plt.imshow(gray_image, cmap="gray", origin="upper")
plt.title("Gambar Asli (Grayscale)")

plt.subplot(2, 3, 5)
plt.imshow(bright_gray_image, cmap="gray", origin="upper")
plt.title("Brightness (Grayscale)")

plt.subplot(2, 3, 6)
plt.imshow(contrast_gray_image, cmap="gray", origin="upper")
plt.title("Contrast (Grayscale)")

plt.tight_layout()
plt.show()
```
